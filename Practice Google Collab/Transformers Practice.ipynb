```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I am extremely flabourgasted for the starting of the third semester")
```

```python
classifier(
    ["I dont like to go to the festival.", "I do not like this so much!"]
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "The Channel that shows the games is the Star Sports",
    candidate_labels=["Channel", "Football", "Cooking"],
)
```

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator(
    "I am very big fan of Cillian Murphy for his acting is just extraordinary",
          max_length=30,
    num_return_sequences=2,

)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("The city of Bengaluru is completely filled with traffic and is almost impossible to find a way out of it")
```
